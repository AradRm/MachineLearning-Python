##Split our needed features to X and Y
##Then use StandardScaler function to bring our features in one scale between 0 - 1
## that mean is "Standardize features by removing the mean and scaling to unit variance."
##at first we make a instance of scaler and fit our X to predict more efficiently

```python
X_Nscale = df_train[['battery_power', 'blue' ,'dual_sim', 'fc', 'four_g', 'int_memory', 'n_cores', 'pc', 'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time', 'three_g', 'wifi']]
```

```python
Y = df_train['price_range']
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X_Nscale)
```

##Now we need to split our X and Y for prediction and train our random forest model
###we import train_test_split from Sklearn.model selection.this function help us to split our data to train and test 
####The parameters that we will use in our spliting are these parameters:
####1-splited= X, Y
####2-train_size:(default=None) If float, should be between 0.0 and 1.0 and represent the proportion of the dataset to include in the train split. If int, represents the absolute number of train samples. If None, the value is automatically set to the complement of the test size.
####random_state:(default=None) Controls the shuffling applied to the data before applying the split. Pass an int for reproducible output across multiple function calls.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X , Y, train_size=0.3, random_state=4)
```

##Now we start build our models and fit our x_train and y_train
###The parameters that we use in our model:
###1-n_stimator:(Default=100) The number of trees in the forest.
###2-max-depth:(Default=None) The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.
###3-criterion{“gini”, “entropy”, “log_loss”}, default=”gini”
###The function to measure the quality of a split. Supported criteria are “gini” for the Gini impurity and “log_loss” and “entropy” both for the Shannon information gain, see Mathematical formulation. Note: This parameter is tree-specific.

```python
model = RandomForestClassifier(n_estimators=300, criterion='gini', max_depth=5)
model.fit(X_train, Y_train)
```

```python
y_prediction = model.predict(X_test)
y_prediction
```

```python
print("accuracy score of our model:", accuracy_score(Y_test, y_prediction))
```

###Now we imposrt classification report:
###Build a text report showing the main classification metrics.

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test, y_prediction))
```

##Now we fit our split data in DecsionTreeClassifire and predict x_testset
###calcute our model accuracy


```python
X_trainset, X_testset, Y_trainset, Y_testset = train_test_split(X , Y, train_size=0.3, random_state=4)
```

```python
predtree = DecisionTreeClassifier(criterion='entropy', max_depth=5)
predtree.fit(X_trainset, Y_trainset)
```
